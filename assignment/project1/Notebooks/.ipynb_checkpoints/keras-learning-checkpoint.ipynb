```python
import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(0,100,30)
y = 3 * x + 7 + np.random.randn(30)*8
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
from keras import layers

model = keras.Sequential()

```

```python
model.add(layers.Dense(1, input_dim=1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss="mse")
```

```python
model.fit(x, y, epochs=3000) #epochs：训练次数
```

```python
model.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x))
```

```python
model.predict([150])
```
